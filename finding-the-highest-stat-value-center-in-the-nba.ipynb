Import Stats of All NBA Players from 2021-2022 Season

```python
import pandas as pd
```

```python
df = pd.read_csv('../input/nba-player-stats/2021-2022 NBA Player Stats - Regular.csv', encoding="Windows-1252", delimiter=";")

df.head()
```

Select all Centers from dataframe

```python
cf=df.loc[(df['Pos']=='C')]
cf
```

Select Centers with 20+ PPG

```python
cfp=cf.loc[(cf['PTS']>20)]
cfp=cfp[['Player','Age','AST','TRB','STL','BLK','PTS']]
cfp
```

Plot As Graph

```python
import matplotlib.pyplot as plt
plt.plot(cfp['Player'],cfp['PTS'])
plt.title('Points per Game Centers')
plt.xlabel('Player')
plt.ylabel('PTS')
for x,y in zip(cfp['Player'],cfp['PTS']):
    label = "{:}".format(x)
    plt.annotate(label,(x,y), textcoords="offset points", xytext=(0,10),ha='center')
ax=plt.gca()
ax.axes.xaxis.set_ticklabels([])
plt.show()
```

Finding common

```python
import copy
dfs_array = [cfa, cfb,cfr,cfp,cfs]
if len(dfs_array) <= 1:
    dfs_array
dfs = copy.copy(dfs_array)
length = len(dfs) 
while length > 1:
    df1 = dfs.pop()
    df2 = dfs.pop()
    df0 = df1.loc[ df1['Player'].isin( df2['Player'].values ) ]
    dfs.insert(0,df0)
    length = len(dfs)
dfs
```
